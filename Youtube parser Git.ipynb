```python
def get_search2(youtube,slovo,Z):
    x=0
    request = youtube.search().list(
                part="snippet",
                maxResults=50,
                regionCode='RU',
                relevanceLanguage='ru',
                type='video',
                q=slovo
            )
    response = request.execute()
    
    channel_ids_search=[]
    
    for i in range(len(response['items'])):
        channel_ids_search.append(response['items'][i]['snippet']['channelId'])
        
    next_page_token = response.get('nextPageToken')
    more_pages = True
    
    while more_pages and x<Z:
        x=x+1
        if next_page_token is None:
            more_pages = False
        else:
            request = youtube.search().list(
                            part="snippet",
                            maxResults=50,
                            regionCode='RU',
                            relevanceLanguage='ru',
                            type='video',
                            pageToken= next_page_token,
                            q=slovo
                        )
            response = request.execute()
                
            for i in range(len(response['items'])):
                channel_ids_search.append(response['items'][i]['snippet']['channelId'])
            
            next_page_token = response.get('nextPageToken')
        
    return channel_ids_search
```

## –ü–æ–ª—É—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –∫–∞–Ω–∞–ª–æ–≤ –∏ —Å–ø–∏—Å–∫–∞ id_playlist —Å –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã–º–∏ –≤–∏–¥–µ–æ

```python
def get_channel_details(youtube, channel_ids):
    
    channel_stats=[]
    playlist_ids=[]
    
    
    for i in range(0, len(channel_ids),50):
        request = youtube.channels().list(
            part='snippet,statistics,contentDetails',
            id=','.join(channel_ids[i:i+50]))
        response = request.execute()
        
        for video in response['items']:
    
            data = dict(Channel_name = video['snippet']['title'],
                        Channel_id = video['id'],
                        Description = video['snippet']['description'],
                        PublishedAt = video['snippet']['publishedAt'],
                        #Country = response['items'][i]['snippet']['country'],
                        Subscribers = video['statistics']['subscriberCount'],
                        Views = video['statistics']['viewCount'],
                        Total_videos = video['statistics']['videoCount'],
                        playlist_id = video['contentDetails']['relatedPlaylists']['uploads'])
        
            playlist_ids.append(video['contentDetails']['relatedPlaylists']['uploads'])
        
            channel_stats.append(data)
    
    return channel_stats, playlist_ids
```

```python
channel_stats,playlist_ids= get_channel_details(youtube, canals_list1)
```

```python
def get_video_ids(youtube, playlist_idis):
    
    video_ids=[]
    x=0
    
    for idi in playlist_idis:
        x=x+1
        print(x)
    
        request = youtube.playlistItems().list(
                    part='contentDetails',
                    playlistId = idi,
                    maxResults = 50)
        response = request.execute()

        

        for i in range(len(response['items'])):
            video_ids.append(response['items'][i]['contentDetails']['videoId'])

        next_page_token = response.get('nextPageToken')
        more_pages = True

        while more_pages:
            if next_page_token is None:
                more_pages = False
            else:
                request = youtube.playlistItems().list(
                            part='contentDetails',
                            playlistId = idi,
                            maxResults = 50,
                            pageToken= next_page_token)
                response = request.execute()

                for i in range(len(response['items'])):
                    video_ids.append(response['items'][i]['contentDetails']['videoId'])

                next_page_token = response.get('nextPageToken')
        
    return video_ids
```

## –ü–æ–ª—É—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –≤–∏–¥–µ–æ

```python
def get_video_details(youtube, video_ids):
    
    all_video_stats=[]
    tags=[]
    commentCount=[]
    likes=[]
    views=[]
    
    for i in range(0, len(video_ids),50):
        request = youtube.videos().list(
            part='snippet,statistics,contentDetails,status',
            id=','.join(video_ids[i:i+50]))
        response = request.execute()
        
        for video in response['items']:
            video_stats = dict(Title = video['snippet']['title'],
                               Description = video['snippet']['description'],
                               Published_date = video['snippet']['publishedAt'],
                               ChannelId = video['snippet']['channelId'],
                              
                               categoryId = video['snippet']['categoryId'],
                               Thumbnails = video['snippet']['thumbnails']['high']['url'],
                               
                               
                               licensedContent=video['contentDetails']['licensedContent'],
                               duration=video['contentDetails']['duration'],
                               definition=video['contentDetails']['definition'],
                               caption=video['contentDetails']['caption'],
                               status=video['status']['madeForKids']
                               
                               )
            
            try:
                tags.append(video['snippet']['tags'])
            
            except KeyError:
                tags.append('–Ω–µ–¥–æ—Å—Ç—É–ø–Ω–æ')
            
            all_video_stats.append(video_stats)
            
            try:
                commentCount.append(video['statistics']['commentCount'])
            
            except KeyError:
                commentCount.append('–Ω–µ–¥–æ—Å—Ç—É–ø–Ω–æ')
                
            try:
                likes.append(video['statistics']['likeCount'])
            
            except KeyError:
                likes.append('–Ω–µ–¥–æ—Å—Ç—É–ø–Ω–æ')
                
            try:
                views.append(video['statistics']['viewCount'])
            
            except KeyError:
                views.append('–Ω–µ–¥–æ—Å—Ç—É–ø–Ω–æ')
            

    video_data = pd.DataFrame(all_video_stats)
    video_data['commentCount']=commentCount
    video_data['tags']=tags
    video_data['likes']=likes
    video_data['views']=views
    
    return video_data
```
